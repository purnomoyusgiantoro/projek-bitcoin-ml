```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/btcusd_1-min_data.csv')
df.head()
```

```python
df.info()
```

```python
f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(), annot=True,linewidths= 0.5, annot_kws={'fontsize': 10}, ax=ax)
plt.show()
```

```python
df.plot(kind='scatter', x='High',y='Low',alpha = 0.5,color = 'red')
plt.xlabel('High')
plt.ylabel('Low')
plt.title('High Low Scatter Plot')
plt.show()
```

```python
df.High.plot(kind='hist',bins= 50,figsize=(10,10))
plt.show()
```
